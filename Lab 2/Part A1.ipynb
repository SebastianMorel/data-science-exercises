```
import matplotlib.pyplot as plt
import numpy as np

%matplotlib inline
```

```
# Generate random numbers
nums = np.round(np.random.random_sample((10, 5)), 4)

# Print data as table
print('Generated data:')
for row in nums:
  for column in row:
    print(column, end='\t')
  print('')
```

```
# 2. Calculate statistics
def calculate_statistics(title, data):
  """
  Calculates and prints statistics.
  
  Parameters
  ----------
  title : string
    Title for the statistics
  
  data : np.matrix
    Matrix of data
  """
  print(title)
  print(f'μ = {data.mean()}')
  print(f'σ = {data.std()}')
  print(f'min = {data.min()}')
  q25 = np.percentile(data, 25)
  q75 = np.percentile(data, 75)
  print(f'first quartile = {q25}')
  print(f'median = {np.median(data)}')
  print(f'third quartile = {q75}')
  print(f'ICQ = {q75 - q25}')
  print(f'max = {np.max(data)}')
```

```
calculate_statistics('Random Data', nums)
```

```
# 3. Create histogram
def make_hist(data, bins):
  """
  Creates and plots histogram using specified bins.

  Parameters
  ----------
  data : np.matrix
    Matrix of data
  
  bins : int
    Number of bins to plot on the histogram
  """
  plt.hist(nums.flatten(), bins=bins)
  plt.title(f'Histogram using {bins} bins')
  plt.show()

make_hist(nums, 8)
make_hist(nums, 5)
```
